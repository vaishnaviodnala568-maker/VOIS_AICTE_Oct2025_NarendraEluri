```python
host_areas = df.groupby(['host name','neighbourhood group'])['calculated host listings count'].max().reset_index()
host_areas.sort_values(by='calculated host listings count', ascending=False).head(5)
```

```python
room_price_area_wise=df.groupby(['neighbourhood group','room type'])['price'].max().reset_index()
room_price_area_wise.sort_values(by='price', ascending=False).head(10)
```

```python
neighbourhood_group=['Bronx','Bronx','Brooklyn','Brooklyn','Staten Island','Manhattan','Queens','Manhattan','Queens','Manhattan']
room_type=['Entire home/apt','Private room','Private room','Entire home/apt','Entire home/apt','Entire home/apt','Entire home/apt','Private room','Entire home/apt','Private room']
room_dict={}
for i in room_type:
  room_dict[i]=room_dict.get(i,0)+1
plt.bar(room_dict.keys(),room_dict.values(),color='green',edgecolor='blue')
plt.title('Room Types')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.show()
```

```python
area_reviews=df.groupby(['neighbourhood group'])['number of reviews'].max().reset_index()
area_reviews
```

```python
area=area_reviews['neighbourhood group']
reviews=area_reviews['number of reviews']
fig=plt.figure(figsize=(10,5))
plt.bar(area,reviews,color="red",width=0.5)
plt.xlabel('Area')
plt.ylabel('Reviews')
plt.title("Number of Reviews in terms of area")
plt.show()
```

```python
price_area=df.groupby(['price'])['number of reviews'].max().reset_index()
price_area.head(10)
```

```python
price_list=price_area['price']
reviews_list=price_area['number of reviews']
fig=plt.figure(figsize=(10,5))
plt.scatter(price_list, reviews_list)
plt.xlabel('price')
plt.ylabel('Number of review')
plt.title('Number of reviews vs price')
plt.show()
```

```python
busy_hosts=df.groupby(['host id','host name','room type'])['number of reviews'].max().reset_index()
busy_hosts=busy_hosts.sort_values(by='number of reviews',ascending=False).head(10)
busy_hosts.head()
```

```python
name_hosts=busy_hosts['host name']
review_got=busy_hosts['number of reviews']
fig=plt.figure(figsize=(10,5))
plt.bar(name_hosts,review_got,color="purple",width=0.5)
plt.xlabel('Host Name')
plt.ylabel('Number of Reviews')
plt.title("Busiest host in terms of reviews")
plt.show()
```

```python
Highest_price=df.groupby(['host id','host name','room type','neighbourhood group'])['price'].max().reset_index()
Highest_price=Highest_price.sort_values(by='price',ascending=False).head(10)
Highest_price.head()
```

```python
name_of_host=Highest_price['host name']
price=Highest_price['price']
fig=plt.figure(figsize=(10,5))
plt.bar(name_of_host,price,color="orange",width=0.5)
plt.xlabel('Host Name')
plt.ylabel('Price')
plt.title("Host with highest price")
plt.show()
```

```python
traffic_areas=df.groupby(['neighbourhood group'])['minimum nights'].count().reset_index()
traffic_areas=traffic_areas.sort_values(by='minimum nights',ascending=False).head(10)
traffic_areas
```

```python
areas_Traffic=traffic_areas['neighbourhood group']
room_stayed=traffic_areas['minimum nights']
fig=plt.figure(figsize=(7,5))
plt.bar(areas_Traffic,room_stayed,color="red",width=0.5)
plt.xlabel("Neighbourhood Group")
plt.ylabel("minimum nights")
plt.title("Traffic Areas Based on Reviews Per Month")
plt.show()
```

```python
numeric_df=df.select_dtypes(include=[np.number])
corr=numeric_df.corr(method='kendall')
fig=plt.figure(figsize=(12,6))
sns.heatmap(corr,annot=True)
numeric_df.columns
```

```python
plt.rcParams['figure.figsize']=[8,6]
ax=sns.countplot(y='room type',hue='neighbourhood group',data=df,palette='bright')
total=len(df['room type'])
for p in ax.patches:
  percentage='{:.1f}%'.format(100*p.get_width()/total)
  x=p.get_width()+0.02
  y=p.get_y()+p.get_height()/2
  ax.annotate(percentage,(x,y),ha='left',va='center')
  plt.title('count of each room types in NYC')
  plt.xlabel('count')
  plt.ylabel('room types')
plt.show()
```
